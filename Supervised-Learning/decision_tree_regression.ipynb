```python
import matplotlib.pyplot as plt
from sklearn.datasets import load_diabetes
from sklearn.model_selection import train_test_split
```

```python
df=load_diabetes(as_frame=True).frame
```

```python
x=df.drop("target",axis=1)
y=df["target"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=.3,random_state=43)
```

```python
from sklearn.tree import DecisionTreeRegressor
model=DecisionTreeRegressor(max_depth=7,min_samples_leaf=20)
```

model.fit(x_train,y_train)

```python
model.fit(x_train,y_train)
```

```python
y_pred_test=model.predict(x_test)
y_pred_train=model.predict(x_train)
```

```python
from sklearn.metrics import r2_score,mean_squared_error
```

```python
print("mse train",mean_squared_error(y_train,y_pred_train))
print("mse tst",mean_squared_error(y_test,y_pred_test))

print("r2_score train",r2_score(y_train,y_pred_train))
print("r2_score_test",r2_score(y_test,y_pred_test))
```

```python
from sklearn.tree import plot_tree

plt.figure(figsize=(10,10))

plot_tree(
    model,
    feature_names=x.columns,
    filled=True
)
plt.tight_layout()
```
